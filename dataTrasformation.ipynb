```python
import pandas as pd

#Lê um arquivo CSV 
df = pd.read_csv(r"C:\Users\xanap\OneDrive\Área de Trabalho\data-transformation-pandas-bi\vg_sales.csv")

# Converte a coluna "Rank" para tipo numérico e substitui a coluna original 
df["Rank"] = pd.to_numeric(df["Rank"])

# As colunas selecionadas são convertidas de milhões de unidades para unidades inteiras
df[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"]] = df[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"]].apply(lambda x: x*1000000)

# Remove as linhas com valores nulos ou inválidos da coluna 'Year' e converte para tipo númerico
df = df.loc[pd.to_numeric(df['Year'], errors = 'coerce').notnull()]

# Converte a coluna 'Year' para o tipo de dados datetime  
df["Year"] = pd.to_datetime(df["Year"], format="%Y").dt.year

# Cria uma nova coluna "Global_Sales" e calcula a soma das vendas de todas as regiões para cada jogo.
df["Global_Sales"] = df["NA_Sales"] + df["EU_Sales"] + df["JP_Sales"] + df["Other_Sales"]

# Converte as colunas selecionada para o tipo inteiro  
df = df.astype({"NA_Sales": "int", "EU_Sales": "int", "JP_Sales": "int", "Other_Sales": "int", "Other_Sales": "int", "Global_Sales": "int"})

print(df)

df.head()
```
